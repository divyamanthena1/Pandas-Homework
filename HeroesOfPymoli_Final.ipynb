```python
import pandas as pd
import csv
import pathlib

csvpath = pathlib.Path('unit_04_pandas_homework_HeroesOfPymoli_Resources_purchase_data.csv')

data_file_df = pd.read_csv(csvpath)


```

Player Count

```python
up_df = data_file_df["SN"].unique()
up_value_df = pd.DataFrame([{"Total Players": len(up_df)}]) # total players
up_value_df
```

Purchasing Analysis (Total)

```python
ui_df = data_file_df["Item ID"].unique()
mean_df = (data_file_df["Price"].mean()) # mean price
purchase_df = (data_file_df["Age"].count()) # total purchases
revenue_df = (data_file_df["Price"].sum()) # total revenue
ui_value_df = pd.DataFrame([{"Number of Unique Items": len(ui_df), 
                             "Average Price": mean_df, 
                             "Number of Purchases": purchase_df, 
                                                        "Total Revenue": revenue_df}])
ui_value_df
```

Gender Demographics

```python
data_unique = data_file_df.drop_duplicates(subset=["SN"]).copy()
bins = [0, 10, 15, 20, 25, 30, 35, 40, 200]
data_unique["Age Range"] = pd.cut(data_unique["Age"], bins, right=False)
count_gender = data_unique["Gender"].value_counts() # count of gender
percentage_gender = (count_gender/data_unique["Gender"].count())*100 # percentage of gender
pd.DataFrame({
    "Total Count": count_gender,
    "Percentage of Players": percentage_gender
})
```

Purchasing Analysis (Gender)

```python
data_purchase_count = data_file_df["Gender"].value_counts() # purchase count
data_purchase_value = data_file_df.groupby("Gender").Price.sum() # average purchase price
pd.DataFrame({
    "Purchase Count": data_purchase_count,
    "Average Purchase Price": data_purchase_value/data_purchase_count,
    "Total Purchase Value": data_purchase_value, # total purchase value
    "Avg Total Purchase per Person": data_purchase_value/data_unique["Gender"].value_counts() # avg total purchase per person
})
```

 Purchasing Analysis (Age)

```python
bins = [0, 10, 15, 20, 25, 30, 35, 40, 200]
data_purchase_age = data_file_df["Age"]
data_file_df["Age Range"] = pd.cut(data_purchase_age, bins, right=False)
data_file_df
data_purchase_count_age = data_file_df["Age Range"].value_counts() # purchase count by age
data_purchase_value_age = data_file_df.groupby("Age Range").Price.sum() # purchase value by age
average_purchase_age = data_purchase_value_age/data_purchase_count_age # average purchase price by age
average_purchase_person_age = (data_purchase_value_age/data_unique["Age Range"].value_counts()) # average total purchase per person

pd.DataFrame({
    "Purchase Count": data_purchase_count_age,
    "Average Purchase Price": average_purchase_age,
    "Total Purchase Value": data_purchase_value_age, 
    "Avg Total Purchase per Person": average_purchase_person_age 
})
```

 Top Spenders

```python
top_purchase_value = data_file_df.groupby("SN").Price.sum().sort_values(ascending = False).head(5) # top purchase value
top_purchase_count = data_file_df.SN.value_counts().sort_values(ascending = False).loc[top_purchase_value.index] # purchase count
top_average_purchase_price = (top_purchase_value/top_purchase_count).sort_values(ascending = False) # average purchase price
pd.DataFrame({
    "Purchase Count": top_purchase_count,
    "Average Purchase Price": top_average_purchase_price,
    "Total Purchase Value": top_purchase_value, 
})
```

Most Popular Items

```python
data_file_df.rename(columns={"Item Name": "Item_Name", "Item ID": "Item_ID"}, inplace = True)
popular_purchase_count = data_file_df.groupby(["Item_ID","Item_Name"]).count()["Price"].sort_values(ascending = False).head(5) # purchase count
top_purchase_idx = data_file_df.groupby(["Item_ID","Item_Name"]).count()["Price"].sort_values(ascending = False).head(5).index
popular_purchase_value = data_file_df.groupby(["Item_ID","Item_Name"]).sum()["Price"].loc[top_purchase_idx] # total purchase value
popular_item_price = popular_purchase_value/popular_purchase_count # item price
pd.DataFrame({
    "Purchase Count": popular_purchase_count,
    "Item Price": popular_item_price,
    "Total Purchase Value": popular_purchase_value, 
})
```

Most Profitable Items

```python
profitable_total_value_desc = data_file_df.groupby(["Item_ID","Item_Name"]).Price.sum().sort_values(ascending = False).head(5) # total purchase value
profitable_purchase_count_desc = data_file_df.groupby(["Item_ID","Item_Name"]).count()["Price"].sort_values(ascending = False).loc[profitable_total_value_desc.index] # purchase count
profitable_item_price_desc = profitable_total_value_desc/profitable_purchase_count_desc # item price
pd.DataFrame({
    "Purchase Count": profitable_purchase_count_desc,
    "Item Price": profitable_item_price_desc,
    "Total Purchase Value": profitable_total_value_desc, 
})
```

Analysis

Three observable trends:

1.) Males tend to purchase significantly more items than males and the age range that purchase the largest number of items is 20 - 24

2.) Although females make far less purchases than males, they spend more on average per person

3.) The two most popular items that were purchased are Final Critic and Oathbreaker, Last Hope of the Breaking Storm and they were purchased 13 and 12 times respectively
